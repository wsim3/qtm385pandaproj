While the experiment has been conducted with two datasets collected to see whether STEM majors earn more than starting median salary by other major, there are some limitations to the results obtained. And it is dangerous to define that non-STEM majors earn less; my interviewees are all on their first year of working and salaries may vary by the workers' job title. Lastly, the fact that I have only collectd 10 samples of my friends and compare them to the starting median salary by each major may bring out biases due to small sample size, and most of the interviewees hold STEM degrees. If this were to be proper study, I would have collected data of bigger sample sizes and more non-STEM degree holders samples and compare them to the actual dataset of salaries by major.
